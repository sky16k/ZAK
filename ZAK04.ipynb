установим следующие критерии выбросов для признаков:
- fixed acidity > 14
- volatile acidity > 1.2
- citric acid > 1
- residual sugar > 30
- chlorides > 0.3
- free sulfur dioxide > 150
- total sulfur dioxide > 350
- density > 1.01
- sulphates > 1.25


```python
df[
(
    (df['fixed acidity'] > 14 )
  | (df['volatile acidity'] > 1.2)
  | (df['citric acid'] > 1)
  | (df['residual sugar'] > 30)
  | (df['chlorides'] > 0.3)
  | (df['free sulfur dioxide'] > 150)
  | (df['total sulfur dioxide'] > 350)
  | (df['density'] > 1.01)
  | (df['sulphates'] > 1.25)
)
].index
```

```python
df.drop( df[
  ( (df['fixed acidity'] > 14 )
  | (df['volatile acidity'] > 1.2)
  | (df['citric acid'] > 1)
  | (df['residual sugar'] > 30)
  | (df['chlorides'] > 0.3)
  | (df['free sulfur dioxide'] > 150)
  | (df['total sulfur dioxide'] > 350)
  | (df['density'] > 1.01)
  | (df['sulphates'] > 1.25)
  )
  ].index
  ).shape[0] / df.shape[0]
```

потери менее 1% - отбрасываем

```python
df = df.drop( df[
  ( (df['fixed acidity'] > 14 )
  | (df['volatile acidity'] > 1.2)
  | (df['citric acid'] > 1)
  | (df['residual sugar'] > 30)
  | (df['chlorides'] > 0.3)
  | (df['free sulfur dioxide'] > 150)
  | (df['total sulfur dioxide'] > 350)
  | (df['density'] > 1.01)
  | (df['sulphates'] > 1.25)
  )
  ].index
  )
df.shape[0]
```
